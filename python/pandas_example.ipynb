```python
# missing data handling
df.isnull().sum()
```

```python
# update missing values in 'Age' column with mean age
mean_age = df['Age'].mean() 
df['Age'].fillna(mean_age, inplace=True)
print(df['Age'].isnull().sum())
```

```python
# Replace cabin values with 'Unknown'
df['Cabin'].fillna('Unknown', inplace=True)
print(df['Cabin'].isnull().sum())
```

```python
# analysis age distribution
import matplotlib.pyplot as plt
plt.hist(df['Age'], bins=20)
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Age Distribution of Titanic Passengers')
plt.show()
```

```python
# analysis age with survival
plt.figure(figsize=(10, 6)) 
plt.hist(df[df['Survived'] == 0]['Age'], bins=20, alpha=0.5, label='Not Survived')
plt.hist(df[df['Survived'] == 1]['Age'], bins=20, alpha=0.5, label='Survived')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Age Distribution by Survival Status')
plt.legend()
```

```python
# analysis age with survival using boxplot
plt.figure(figsize=(10, 6))
df.boxplot(column='Age', by='Survived')
plt.xlabel('Survived')
plt.ylabel('Age')
plt.title('Age Distribution by Survival Status')
plt.suptitle('')
plt.show()
```

```python
# survival rate by age group
df['AgeGroup'] = pd.cut(df['Age'], bins=[0, 12, 18, 35, 60, 100], labels=['Child', 'Teen', 'Adult', 'Middle Age', 'Senior'])
survival_rate = df.groupby('AgeGroup')['Survived'].mean()
print(survival_rate)    
```

```python
# survival with fare analysis
plt.figure(figsize=(10, 6))
plt.hist(df[df['Survived'] == 0]['Fare'], bins=20,
            alpha=0.5, label='Not Survived')
plt.hist(df[df['Survived'] == 1]['Fare'], bins=20,
            alpha=0.5, label='Survived')
plt.xlabel('Fare')
plt.ylabel('Frequency')
plt.title('Fare Distribution by Survival Status')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.hist(df[df['Age'].notnull()]['Age'], bins=5)
plt.axvline(x=df['Age'].mean(), color='red', linestyle='--', label='Mean Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Age Distribution of Titanic Passengers')
plt.show()
```
