```python
## Lambad function
greating = lambda name: f"Hello, {name}!"
print(greating("Alice"))
```

```python
## Map function
numbers = [1, 2, 3, 4, 5]
squared_numbers = list(map(lambda x: x**2, numbers))
print(squared_numbers)
```

```python
## List comprehension
list_comprehension = [x**2 for x in numbers]
print(list_comprehension)
```

```python
## filiter function
even_numbers = list(filter(lambda x: x % 2 == 0, numbers))
print(even_numbers)
```

```python
class Employee:
    def __init__(self, name, salary):
        self.name = name
        self.salary = salary

    def __repr__(self):
        return f"Employee(name={self.name}, salary={self.salary})"
```

```python
employees = [
    Employee("Alice", 70000),
    Employee("Bob", 50000),
    Employee("Charlie", 60000)
]
print(employees)
sorted_employees = sorted(employees, key=lambda emp: emp.salary)
print(sorted_employees)
```

```python
### Error handling
try:        
    result = 10 / 0
except ZeroDivisionError as e:
    print(f"Error: {e}")
```

```python
### Exception handling with multiple except blocks
try:
    num = int(input("Enter a number: "))
    result = 10 / num
except ValueError:
    print("Invalid input. Please enter a valid number.")
except ZeroDivisionError:
    print("Error: Division by zero is not allowed.")
except Exception as e:
    print(f"An unexpected error occurred: {e}")
```

```python
# Write a Python program that uses global and local variables to perform a calculation. Print the values of both the global and local variables before and after the calculation.   
global_var = 10
def calculate():
    local_var = 5
    print(f"Inside function - Global variable: {global_var}, Local variable: {local_var}")
    result = global_var + local_var
    print(f"Result of calculation: {result}")

print(f"Before function call - Global variable: {global_var}")
calculate()
print(f"After function call - Global variable: {global_var}")
```

```python
# Write a Python program to sort a list of strings based on the length of each string. Define a custom sort function that takes a list of strings and returns a sorted list.
def sort_by_length(strings):
    return sorted(strings, key=lambda s: len(s))

string_list = ["apple", "banana", "kiwi", "grapefruit", "fig"]
sorted_strings = sort_by_length(string_list)    
print(sorted_strings)
```

```python
#  Write a Python program to find the longest word in a file.
def find_longest_word(file_path):
    longest_word = ""
    with open(file_path, "r") as file:
        for line in file:
            words = line.split()
            for word in words:
                if len(word) > len(longest_word):
                    longest_word = word
    return longest_word

longest = find_longest_word("test.txt")
print(f"The longest word in the file is: {longest}")
```
