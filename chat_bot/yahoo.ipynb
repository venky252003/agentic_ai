```python
import requests

API_KEY = os.getenv("ALPHA_API_KEY")
symbol = "AAPL"

url = f"https://www.alphavantage.co/query?function=TIME_SERIES_DAILY&symbol={symbol}&apikey={API_KEY}"
response = requests.get(url)
data = response.json()

print(data)
```

```python
import json
from datetime import datetime

# Extract date-wise trade history from the API response
def extract_trade_history(api_data):
    """
    Extract date-wise trade history from Alpha Vantage API response
    Returns a JSON object with structured trade data
    """
    trade_history = {
        "meta_data": {},
        "time_series": []
    }
    
    # Extract metadata
    if "Meta Data" in api_data:
        meta = api_data["Meta Data"]
        trade_history["meta_data"] = {
            "symbol": meta.get("2. Symbol", ""),
            "last_refreshed": meta.get("3. Last Refreshed", ""),
            "interval": meta.get("4. Interval", ""),
            "output_size": meta.get("5. Output Size", ""),
            "time_zone": meta.get("6. Time Zone", "")
        }
    
    # Extract time series data
    if "Time Series (Daily)" in api_data:
        time_series = api_data["Time Series (Daily)"]
        
        for date, values in time_series.items():
            trade_record = {
                "date": date,
                "open": float(values.get("1. open", 0)),
                "high": float(values.get("2. high", 0)),
                "low": float(values.get("3. low", 0)),
                "close": float(values.get("4. close", 0)),
                "volume": int(values.get("5. volume", 0))
            }
            trade_history["time_series"].append(trade_record)
    
    # Sort by date (most recent first)
    trade_history["time_series"].sort(key=lambda x: x["date"], reverse=True)
    
    # Add summary statistics
    if trade_history["time_series"]:
        closes = [t["close"] for t in trade_history["time_series"]]
        volumes = [t["volume"] for t in trade_history["time_series"]]
        
        trade_history["summary"] = {
            "total_records": len(trade_history["time_series"]),
            "highest_close": max(closes),
            "lowest_close": min(closes),
            "average_close": sum(closes) / len(closes),
            "total_volume": sum(volumes)
        }
    
    return trade_history

# Process the data and create JSON object
trade_data = extract_trade_history(data)

# Display results
print("=== TRADE HISTORY ===")
print(f"Symbol: {trade_data['meta_data'].get('symbol')}")
print(f"Last Refreshed: {trade_data['meta_data'].get('last_refreshed')}")
print(f"\nTotal Records: {trade_data['summary'].get('total_records')}")
print(f"Highest Close: ${trade_data['summary'].get('highest_close'):.2f}")
print(f"Lowest Close: ${trade_data['summary'].get('lowest_close'):.2f}")
print(f"Average Close: ${trade_data['summary'].get('average_close'):.2f}")
print(f"Total Volume: {trade_data['summary'].get('total_volume'):,}")

print("\n=== RECENT TRADES (First 5 Days) ===")
for record in trade_data['time_series'][:5]:
    print(f"Date: {record['date']}, Close: ${record['close']:.2f}, Volume: {record['volume']:,}")

# Save to JSON file
with open('trade_history.json', 'w') as f:
    json.dump(trade_data, f, indent=2)
    print("\n✓ Trade history saved to 'trade_history.json'")

# Display full JSON
print("\n=== FULL JSON OBJECT ===")
print(json.dumps(trade_data, indent=2)[:500] + "...")  # Show first 500 chars
```

```python
# Extract dates and prices in the requested format
def extract_dates_and_prices(trade_data):
    """
    Extract dates and close prices in simple format
    Returns: {"dates": dates, "prices": prices}
    """
    dates = [record["date"] for record in trade_data["time_series"]]
    prices = [record["close"] for record in trade_data["time_series"]]
    
    return {
        "dates": dates,
        "prices": prices
    }

# Get the formatted data
result = extract_dates_and_prices(trade_data)

# Display the result
print("=== DATES AND PRICES ===")
print(json.dumps(result, indent=2))

# Save to JSON file
with open('dates_prices.json', 'w') as f:
    json.dump(result, f, indent=2)
    print("\n✓ Saved to 'dates_prices.json'")

# Quick view
print("\n=== SAMPLE DATA ===")
for i in range(5):
    print(f"{result['dates'][i]}: ${result['prices'][i]:.2f}")
```
