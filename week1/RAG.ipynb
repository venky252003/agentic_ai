```python
from langchain_core.documents import Document

documents = [
    Document(page_content="‘Help is on its way’: Donald Trump freezes Iran talks, calls on protesters to ‘take over institutions’", metadata={"source": "moneycontrol_news"}),
    Document(page_content="""Trade setup for January 14: Top 15 things to know before the opening bell
           Flipkart, Fidelity, TPG-backed logistics platform Shadowfax to launch Rs 1,907-crore IPO on January 20; listing scheduled on January 28""",
          metadata={"source": "mint_news"}),
    Document(page_content="Stocks to Watch Today: TCS, HCL Tech, Anand Rathi Wealth, Biocon, Sical Logistics, NLC India, Gabion Technologies in focus on 13 January", metadata={"source": "moneycontrol_news"}),
    Document(page_content="India's retail inflation eases to 5.72% in December; industrial output grows 4.3% in November", metadata={"source": "ndtv_news"}),
    Document(page_content="Sensex, Nifty likely to open flat; HDFC Bank, ICICI Bank, TCS, Infosys to lead gains", metadata={"source": "ibnlive_news"}),
]
```

```python
from langchain_chroma import Chroma
from langchain_huggingface import HuggingFaceEmbeddings
import tqdm as notebook_tqdm

embeddings = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")
vectorstore = Chroma.from_documents(documents, embeddings)
```

```python
results = vectorstore.similarity_search("What is the latest news on stocks?", k=2)
for i, doc in enumerate(results):
    print(f"Result {i+1}:\n{doc.page_content}\nSource: {doc.metadata['source']}\n")
```

```python
### Similarity Search with Relevance Scores
results_with_scores = vectorstore.similarity_search_with_score("What is the latest news on stocks?", k=2)
for i, (doc, score) in enumerate(results_with_scores):
    print(f"Result {i+1} (Score: {score:.4f}):\n{doc.page_content}\nSource: {doc.metadata['source']}\n")
```
