```python
# Load environment variables from .env file
from dotenv import load_dotenv
import os

load_dotenv()

# Get OpenAI API key from environment
openai_api_key = os.getenv("OPENAI_API_KEY")
if not openai_api_key:
    print("Warning: OPENAI_API_KEY not found in .env file")
else:
    print("✓ OpenAI API key loaded successfully")
```

```python
# Import LangChain components
from langchain_openai import ChatOpenAI
from langchain.prompts import ChatPromptTemplate
from langchain.schema import HumanMessage

# Initialize OpenAI LLM with ChatOpenAI
llm = ChatOpenAI(
    model="gpt-3.5-turbo",
    temperature=0.7,
    api_key=openai_api_key
)

print("✓ LangChain ChatOpenAI model initialized successfully")
```

```python
# Test 1: Simple message to the model
print("=" * 50)
print("Test 1: Simple Query")
print("=" * 50)

response = llm.invoke("What is the capital of France?")
print(response.content)
```

```python
# Test 2: Using ChatPromptTemplate
print("\n" + "=" * 50)
print("Test 2: Using Prompt Template")
print("=" * 50)

prompt = ChatPromptTemplate.from_messages([
    ("system", "You are a helpful AI assistant."),
    ("user", "Tell me a fun fact about {topic}")
])

chain = prompt | llm
response = chain.invoke({"topic": "Python programming"})
print(response.content)
```

```python
# Test 3: Streaming responses (optional)
print("\n" + "=" * 50)
print("Test 3: Streaming Response")
print("=" * 50)

stream = llm.stream("Write a short haiku about artificial intelligence")
print("".join(chunk.content for chunk in stream))
```
